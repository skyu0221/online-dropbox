一、Numpy
=========
[Numpy](https://numpy.org/devdocs/user/quickstart.html) 是 Python 中最常使用的一个针对矩阵或维度数组进行运算的 module。Python 本身对 list 等的运算很慢，而且不支持将 list 视为一个整体的运算。Numpy 则提供了一个快速简单的算法。Numpy 可以说类似于 Python 版的 Matlab。

````python
import numpy as np  # 通常大家都将 numpy 改名为 np 来方便输入
````

Numpy 的核心是它的一个 class：[ndarray](https://numpy.org/doc/1.18/reference/generated/numpy.ndarray.html)

[np.ndarray](https://numpy.org/doc/1.18/reference/generated/numpy.ndarray.html) 包含以下几种重要的 attributes：

+ **T**: 矩阵的 transpose (翻转矩阵)
+ **dtype**: 矩阵中数据的 type
+ **flat**: 转成 1D 后的矩阵
+ **size**: 矩阵中的 object 的个数
+ **ndim**: 矩阵的维度
+ **shape**: 矩阵每个维度上有多长

### 创建一个 [np.ndarray](https://numpy.org/doc/1.18/reference/generated/numpy.ndarray.html)

我们可以选择将已有的 list 转换成 [ndarray](https://numpy.org/doc/1.18/reference/generated/numpy.ndarray.html)，或者使用 numpy 的 function 生成一些特殊的 [ndarray](https://numpy.org/doc/1.18/reference/generated/numpy.ndarray.html)

#### 从 list 转换
[`np.array(要转换的 list, dtype=想要的格式)`](https://numpy.org/doc/1.18/reference/generated/numpy.array.html)

#### 从 numpy 直接生成
[`np.zeros(大小)`](https://numpy.org/doc/1.18/reference/generated/numpy.zeros.html) 生成一个给定大小的，全是 0 的矩阵

[`np.ones(大小)`](https://numpy.org/doc/1.18/reference/generated/numpy.ones.html) 生成一个给定大小的，全是 1 的矩阵

[`np.empty(大小)`](https://numpy.org/doc/1.18/reference/generated/numpy.empty.html)  生成一个给定大小，但是不知道里面的值是什么的矩阵

[`np.arange(开始，截止，变化值)`](https://numpy.org/doc/1.18/reference/generated/numpy.arange.html) 生成一个类似于 range 函数结果的 1D 矩阵

[`np.linspace(开始，截止，段数)`](https://numpy.org/doc/1.18/reference/generated/numpy.linspace.html) 生成一个 1D 等分矩阵

[`np.eye(大小)`](https://numpy.org/doc/1.18/reference/generated/numpy.eye.html) 生成一个 2D 的 identity 矩阵

[`np.full(大小, 值)`](https://numpy.org/doc/1.18/reference/generated/numpy.full.html) 生成一个给定大小的，全是指定值的矩阵

[`np.random.random(大小)`](https://numpy.org/doc/1.18/reference/random/generated/numpy.random.random.html) 生成一个给定大小的，全是 \[0, 1) 随机值的矩阵

**注意：一个标准的矩阵应当每个维度上长度恒定，如果不同的话会造成很多麻烦**

### np.array 的基础运算

[np.ndarray](https://numpy.org/doc/1.18/reference/generated/numpy.ndarray.html) 支持多种不同的矩阵与矩阵间运算，也支持 element-wise 的矩阵与单独值的运算。

#### Element-wise operation

##### Calculation

+ `+`: 当矩阵与矩阵相加时，两个矩阵必须大小相同，相同位置的值相加
+ `-`: 原理与加法相同
+ `*`: 矩阵可以和一个数字相乘，或者和一个矩阵相乘。当和矩阵相乘时，相同位置的值相乘
+ `/`: 原理与除法相同
+ `**`: 次方运算
+ [`np.sin(A)`](https://numpy.org/doc/1.18/reference/generated/numpy.sin.html): 对每个 element 计算 sin 值 (假设提供的值是 radius)
+ [`np.exp(A)`](https://numpy.org/doc/1.18/reference/generated/numpy.exp.html): 对每个 element 计算 expoenetial 值 (e^A)
+ [`np.sqrt(A)`](https://numpy.org/doc/1.18/reference/generated/numpy.sqrt.html): 对每个 element 计算开根号

##### Comparison
+ `>=`, `<=`, `==`, `!=`, `<`, `>`: 当矩阵与矩阵相比较的时候，相同位置的值进行比较

````python
# Create a numpy array [1 2 3 4], float type
a = np.array([1, 2, 3, 4])
# Create a numpy array [10 20 30 40], float type
b = np.array([10, 20, 30, 40])
print(f"a:\n{a}\nb:\n{b}")
````

````python
print(f"a + b:\n{a + b}\n")
print(f"a - b:\n{a - b}\n")
print(f"a * b:\n{a * b}\n")
print(f"a * 4:\n{a * 4}\n")
print(f"a / b:\n{a / b}\n")
print(f"b ** a:\n{b ** a}")
````

#### Matrix operation

##### 矩阵与矩阵间运算

+ [`np.dot(A, B)`](https://numpy.org/doc/1.18/reference/generated/numpy.dot.html): 矩阵点乘

##### 矩阵整体运算

+ [`np.transpose(A)`](https://numpy.org/doc/1.18/reference/generated/numpy.transpose.html): 矩阵置换
+ [`np.linalg.inv(A)`](https://numpy.org/doc/1.18/reference/generated/numpy.linalg.inv.html): 矩阵 inverse

##### 矩阵内部运算

进行内部运算时，一般需要我们提供一个运算的方向 `axis=?`。如果不加则把整个 matrix 视为一个整体。

如果 `axis=0`，那么就表示运算应用在每一列上。如果 `axis=1`，那么就表示运算应用在每一行上。

+ [`np.sum(A, axis=?)`](https://numpy.org/doc/1.18/reference/generated/numpy.sum.html): 对每一行或每一列进行求和
+ [`np.diff(A, axis=?)`](https://numpy.org/doc/1.18/reference/generated/numpy.diff.html): 对每一行或每一列求后数减前数的差
+ [`np.min(A, axis=?)`](https://numpy.org/doc/1.18/reference/generated/numpy.min.html): 每一行或每一列的最小值
+ [`np.max(A, axis=?)`](https://numpy.org/doc/1.18/reference/generated/numpy.max.html): 每一行或每一列的最大值
+ [`np.median(A, axis=?)`](https://numpy.org/doc/1.18/reference/generated/numpy.median.html): 每一行或每一列的中位数
+ [`np.mean(A, axis=?)`](https://numpy.org/doc/1.18/reference/generated/numpy.mean.html): 对每一行或每一列进行求平均
+ [`np.var(A, axis=?)`](https://numpy.org/doc/1.18/reference/generated/numpy.var.html): 对每一行或每一列进行求 variance
+ [`np.cumsum(A, axis=?)`](https://numpy.org/doc/1.18/reference/generated/numpy.cumsum.html): 对每一行或每一列进行累加
+ [`np.prod(A, axis=?)`](https://numpy.org/doc/1.18/reference/generated/numpy.prod.html): 对每一行或每一列进行累乘
+ [`np.argmin(A, axis=?)`](https://numpy.org/doc/1.18/reference/generated/numpy.argmin.html): 每一行或每一列的最小值所在位置
+ [`np.argmax(A, axis=?)`](https://numpy.org/doc/1.18/reference/generated/numpy.argmax.html): 每一行或每一列的最大值所在位置

+ `x[1,2,...]` 等效于 `x[1,2,:,:,:]`
+ `x[...,3]` 等效于 `x[:,:,:,:,3]`
+ `x[4,...,5,:]` 等效于 `x[4,:,:,5,:]`

Boolean values 可以被用作 index!

```

i = np.array([[True, False], [False, True]])
```

二、Matplotlib
============

[Matplotlib](https://matplotlib.org/api/index.html) 是目前 Python 中最为广泛应用的画图工具。

````python
import matplotlib.pyplot as plt  # 通常都将这个 module 称为 plt，用来减少输入的长度
````

我们所最常使用的 matplotlib 的 sub-module 叫做 [matplotlib.pyplot](https://matplotlib.org/api/pyplot_summary.html)。这个课程只会介绍这一个 sub-module，因为其他的 sub-module 对 ML 而言完全没有用途。

### 画图像
+ `plt.subplot()`: 在一张图片中包含多张小图
+ `plt.plot()`: 曲线图
+ `plt.scatter()`: 散点图
+ `plt.fill_between()`: 在两条 function 中间填满
+ `plt.hist()`: 柱状图
+ `plt.bar()`: 柱状图 (柱之间有间隔)
+ `plt.imshow()`: 显示图片

#### 各类图中重要的参数

+ `color=颜色`: 这个图的颜色
+ `label=名称`: 这个图的图示中名称
+ `alpha=透明度`: 这个颜色的透明度
+ `marker=形状`: 数据点的形状
+ `markersize=大小`: 数据点的大小
+ `s=大小`: 散点图数据点的大小
+ `bins=格数`: Histogram 中分成多少个 bin


### 图像上添加信息

+ `plt.figure()`: 图像设定
+ `plt.xlabel(名称)`： x 轴名称
+ `plt.ylabel(名称)`: y 轴名称
+ `plt.xlim(范围)`: 把 x 轴限制在某个范围
+ `plt.ylim(范围)`: 把 y 轴限制在某个范围
+ `plt.legend()`: 显示图例
+ `plt.title(标题)`: 显示标题
+ `plt.show()`: 显示图片

````python
# Practice!
x = np.linspace(0, 10, 20)
y1 = np.sin(np.linspace(0, 10, 20))
y2 = np.cos(np.linspace(0, 10, 20))
values = np.random.random((20,)) * 10
plt.hist(values, color="orange", label="random", alpha=0.3)
plt.plot(x, y1, color='r', marker='+', label="Sin function", markersize=20)
plt.scatter(x, y2, color='blue', marker='*', label="Cos function", s=1000)
plt.fill_between(x, y1, 0, color='green', alpha=0.3)
plt.xlabel("Population")
plt.ylabel("Temperature")
plt.title("What is the relation between population and temperature?")
plt.ylim([-1, 1])
plt.xlim([0, 3.14])
plt.legend()
plt.show()
````

# 实战演练

````python
from skimage import data
import matplotlib.pyplot as plt

image = data.astronaut()

plt.imshow(image)
plt.show()
````

````python
plt.imshow(image[:,:,0], cmap='gray')
plt.show()
````

````python
plt.imshow(image[:,:,1], cmap='gray')
plt.show()
````

````python
transform = np.array([[0.5, -0.4],
                     [1.2,  2.3]])

row_one = []
for i in range(image.shape[0]):
    row_one += [i] * image.shape[1]


row_two = list(range(image.shape[1])) * image.shape[0]

row_one = np.array(row_one)
row_two = np.array(row_two)

original = np.vstack([row_one, row_two])

transform = np.dot(transform, original).astype(int)
print(transform)

transform[0, :] -= transform[0, :].min()
transform[1, :] -= transform[1, :].min()

new_image = np.zeros(list(transform.max(axis=1) + 1) + [3])

for i in range(np.prod(image.shape[0:2])): # [0:2] -> 要[0], [1]
    new_image[transform[0, i], transform[1, i], :] = image[original[0, i], original[1, i], :] 

# (4, 5), (3, 4), (2, 8) -> rotate 90 degree
# T * [[0, 0, 0, 1, 1, 1],
#      [0, 1, 2, 0, 1, 2]]
````

````python
plt.figure(figsize=(10, 20))

plt.subplot(121)
plt.title("Astronaut after transform")
plt.ylabel("Y")
plt.imshow(new_image.astype(int))
plt.subplot(122)
plt.imshow(image)
plt.plot(values * 40)
plt.title("Astronaut transform")
plt.show()
````
